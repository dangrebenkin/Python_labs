```python
import numpy as np
import math
from sympy import symbols, Eq, solve
```

```python
k = 2
n = -1
```

```python
def task_1():
    A = np.array([n+k, 2*n, -k-n])
    B = np.array([n-k, 2*n+k, k-n])
    C = np.array([n+2*k, 2*n-2*k, k-n])
    ab = B - A
    ac = C - A
    ab_ac_multiplication = np.cross(ab, ac)
    triangle_square = 0.5 * math.sqrt(np.sum(np.power(ab_ac_multiplication, 2))) # площадь треугольника
    parallelogram_square = 2 * triangle_square # площадь параллелограмма
    ac_length = math.sqrt(np.sum(np.power(ac, 2)))
    h = parallelogram_square / ac_length
    return int(triangle_square), int(h)


def task_2():
    A = np.array([n+k, 2*n, -k-n])
    B = np.array([n-k, 2*n+k, k-n])
    C = np.array([n+2*k, 2*n-2*k, k-n])
    D = np.array([3*k + n, 2*k + 2*n, -n])
    ab = B - A
    ac = C - A
    ad = D - A
    v_matrix = np.array([ab, ac, ad])
    v_matrix_det = abs(math.ceil(np.linalg.det(v_matrix))) # объем параллелепипеда
    abcd_volume = v_matrix_det / 6 # объем пирамиды

    # из предыдущей задачи
    ab_ac_multiplication = np.cross(ab, ac)
    triangle_square = 0.5 * math.sqrt(np.sum(np.power(ab_ac_multiplication, 2)))
    parallelogram_square = 2 * triangle_square

    de = v_matrix_det / parallelogram_square # высота пирамиды
    return int(abcd_volume), int(de)


def task_3(y_F: int):
    x, y, z = symbols('x, y, z')

    # уравнение 1
    m1x_1 = k**2
    m2y_1 = -3*k
    m3z_1 = -n
    v_1 = 3*(k**2)*n
    equation_1 = Eq(m1x_1*x + m2y_1*y + m3z_1*z + v_1, 0)

    # уравнение 2
    m1x_2 = k
    m2y_2 = -2
    v_2 = k*n
    equation_2 = Eq(m1x_2*x + m2y_2*y + v_2, 0)

    # выбираем y_F
    equation_3 = Eq(y, y_F)
    solution = solve([equation_1, equation_2, equation_3], 
                       (x, y, z))
    
    n_1 = np.array([m1x_1, m2y_1, m3z_1])
    n_2 = np.array([m1x_2, m2y_2, 0])
    n_1_n_2_multiplication = np.cross(n_1, n_2)

    # канонические уравнения
    canonical_equations = f'x-({solution[x]})/{n_1_n_2_multiplication[0]} ' \
        f'= y-({solution[y]})/{n_1_n_2_multiplication[1]} = z-({solution[z]})/{n_1_n_2_multiplication[2]}'
    return canonical_equations
    

def task_4():
    M = np.array([-2*(k**2), 2*k*n, -(n*k)])
    x, y, z, t = symbols('x, y, z, t')
    m1x = 2*k
    m2y = -3*n
    m3z = n
    v = -3*(n**2)*k
    main_equation = Eq(m1x*x + m2y*y + m3z*z + v, 0)
    equation_x = Eq(M[0] + m1x*t, x)
    equation_y = Eq(M[1] + m2y*t, y)
    equation_z = Eq(M[2] + m3z*t, z)
    solution = solve([main_equation, 
                       equation_x, 
                       equation_y, 
                       equation_z], 
                       (x, y, z, t))
    return np.array([solution[x], solution[y], solution[z]]).tolist()


def task_5():
    
    # угол
    l1_denominators = np.array([2*k + n, 2*k - 2*n, k + 2*n]) 
    l2_denominators = np.array([k - 2*n, -2*n - 2*k, 2*k - n])
    magnitude_l1 = np.linalg.norm(l1_denominators)
    magnitude_l2 = np.linalg.norm(l2_denominators)
    angle_cos = np.dot(l1_denominators, l2_denominators) / (magnitude_l1 * magnitude_l2)
    angle = np.degrees(np.arccos(angle_cos)).item()
    if angle >= 90:
        angle = 180 - angle

    # расстояние
    l1_numerators = np.array([2*k, 2*k - 4*n, n]) * (-1)
    l2_numerators = np.array([-k - 3*n, -k - 4*n, 3*k - 2*n]) * (-1)
    m1m2 = l2_numerators - l1_numerators
    q1q2_m1m2_matrix = np.array([l1_denominators, l2_denominators, m1m2])
    q1q2_m1m2_matrix_det = abs(math.ceil(np.linalg.det(q1q2_m1m2_matrix)))
    q1q2_multiplication = np.cross(l1_denominators, l2_denominators)
    p_l1l2 = q1q2_m1m2_matrix_det / math.sqrt(np.sum(np.power(q1q2_multiplication, 2)))
    return angle, p_l1l2
```
